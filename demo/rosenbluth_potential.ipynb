# Potential Reconstruction in the Cylindrical Coordinate

```python
import torch
from pyapes.core.geometry import Cylinder
from pyapes.core.mesh import Mesh

from pyrfp.training_data import RosenbluthPotentials_RZ, analytic_potentials_rz_cpu


mesh = Mesh(Cylinder[0:5, -5:5], None, [32, 64])

mnts_eq = torch.tensor(
    [1, 0, 1, 0, 3, 2, 8, 0, 2], dtype=mesh.dtype.float, device=mesh.device
)

RP_rz = RosenbluthPotentials_RZ(
    mesh,
    optimizer_config={
        "lr": 1.0,
        "max_iter": 200,
        "gtol": 1e-6,
        "xtol": 1e-6,
    },
    solver_config={
        "method": "bicgstab",
        "tol": 1e-8,
        "max_it": 1000,
        "report": False,
    },
)

pots_mnts = RP_rz.from_moments(mnts_eq)["pots"]

assert pots_mnts is not None

H_exact = analytic_potentials_rz_cpu(mesh.grid, mesh.grid, pots_mnts["pdf"], "H")
G_exact = analytic_potentials_rz_cpu(mesh.grid, mesh.grid, pots_mnts["pdf"], "G")

jac_H_exact = torch.gradient(H_exact, spacing=mesh.dx.tolist(), edge_order=2)
jac_G_exact = torch.gradient(G_exact, spacing=mesh.dx.tolist(), edge_order=2)
hess_G_exact = torch.gradient(jac_G_exact[0], spacing=mesh.dx.tolist(), edge_order=2) # for rr and rz not zz

from pymyplot import myplt as plt
from pymyplot.colors import TOLCmap

_, axes = plt.subplots(2, 6, figsize=(30/2.54, 12/2.54))
axes[0][0].plot(mesh.grid[0][:, 16], H_exact[:, 16], "ro", label="exact")
axes[0][0].plot(mesh.grid[0][:, 16], pots_mnts["H"][:, 16], "b-", label="computed")

axes[1][0].plot(mesh.grid[0][:, 16], G_exact[:, 16], "ro")
axes[1][0].plot(mesh.grid[0][:, 16], pots_mnts["G"][:, 16], "b-")

axes[0][1].contourf(mesh.grid[0], mesh.grid[1], H_exact, 20, cmap=TOLCmap.sunset())
axes[1][1].contourf(mesh.grid[0], mesh.grid[1], pots_mnts["H"], 20, cmap=TOLCmap.sunset())

axes[0][2].contourf(mesh.grid[0], mesh.grid[1], G_exact, 20, cmap=TOLCmap.sunset())
axes[1][2].contourf(mesh.grid[0], mesh.grid[1], pots_mnts["G"], 20, cmap=TOLCmap.sunset())

axes[0][3].contourf(mesh.grid[0], mesh.grid[1], jac_H_exact[0], 20, cmap=TOLCmap.sunset())
axes[1][3].contourf(mesh.grid[0], mesh.grid[1], pots_mnts["jacH"].r, 20, cmap=TOLCmap.sunset())

axes[0][4].contourf(mesh.grid[0], mesh.grid[1], jac_G_exact[0], 20, cmap=TOLCmap.sunset())
axes[1][4].contourf(mesh.grid[0], mesh.grid[1], pots_mnts["jacG"].r, 20, cmap=TOLCmap.sunset())

axes[0][5].contourf(mesh.grid[0], mesh.grid[1], hess_G_exact[0], 20, cmap=TOLCmap.sunset())
axes[1][5].contourf(mesh.grid[0], mesh.grid[1], pots_mnts["hessG"].rr, 20, cmap=TOLCmap.sunset())

plt.show()
```
